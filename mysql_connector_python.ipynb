```python
#--->OPTIONAL
#Convert Seats_Available to replace
df_final["Seats_Available"] = df_final["Seats_Available"].str.replace("Seats available","")
df_final["Seats_Available"] = df_final["Seats_Available"].str.strip()
#df_final

# Ensure 'Seats_Available' is numeric, converting invalid entries to NaN
df_final['Seats_Available'] = pd.to_numeric(df_final['Seats_Available'], errors='coerce').astype('Int64')

#--->Using .loc[] with inplace=True:
                #Use .loc[] to avoid potential chained assignment issues, though inplace=True isn't needed in this case:
#df_final.loc[:, 'Seats_Available'] = df_final['Seats_Available'].fillna(0)
df_final
```
