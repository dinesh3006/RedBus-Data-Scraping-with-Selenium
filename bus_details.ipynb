```python
from selenium import webdriver
#from selenium.webdriver import AcrionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

#retrieve the bus details 
driver_APSRTC=webdriver.Chrome()
driver_APSRTC.maximize_window()
Bus_Names_APSRTC=[]
Bus_Types_APSRTC=[]
Departure_Time_APSRTC=[]
Destination_Time_APSRTC=[]
Total_Duration_APSRTC=[]
Ratings_APSRTC=[]
Prices_APSRTC=[]
Seats_Available_APSRTC=[]
Route_names=[]
Route_Links=[]

for i,r in df.iterrows():
    link_APSRTC=r["Bus Routes Link"]
    Route_APSRTC=r["Bus Routes Name"]
    
   
#Loop through each link
    driver_APSRTC.get(link_APSRTC)
    time.sleep(2)

    #try:
        # Click elements to reveal bus details
    elements = driver_APSRTC.find_elements(By.XPATH, f"//a[contains(@href, '{link_APSRTC}')]")
    for element in elements:
            element.click()
            time.sleep(2)
    #except:
        #continue

    #   Click elements to views bus
    try:
        clicks=driver_APSRTC.find_element(By.XPATH,"//div[@class='button']")
        clicks.click()
    except:
        continue
    time.sleep(2)

    scrolling = True
    while scrolling:
        old_page_source = driver_APSRTC.page_source

        #Use Actions to perform a PAGE_DOWN
        ActionChains(driver_APSRTC).send_keys(Keys.PAGE_DOWN).perform()

        time.sleep(0.5)

        new_page_source=driver_APSRTC.page_source

        if new_page_source == old_page_source:
            scrolling = False


    #Extract bus details
    Bus_Names=driver_APSRTC.find_elements(By.XPATH,"//div[@class='travels lh-24 f-bold d-color']")
    Bus_Types=driver_APSRTC.find_elements(By.XPATH,"//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    Departure_Time=driver_APSRTC.find_elements(By.XPATH,"//div[@class='dp-time f-19 d-color f-bold']")
    Destination_Time=driver_APSRTC.find_elements(By.XPATH,"//div[@class='bp-time f-19 d-color disp-Inline']")
    Total_Duration=driver_APSRTC.find_elements(By.XPATH,"//div[@class='dur l-color lh-24']")
    
    try:
        Ratings=driver_APSRTC.find_elements(By.XPATH,"//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    except:    
        continue
    Prices=driver_APSRTC.find_elements(By.XPATH,"//div[@class='fare d-block']")
    Seats_Available=driver_APSRTC.find_elements(By.XPATH,"//div[contains(@class,'seat-left')]")

    #Append data to respective lists
    for bus in Bus_Names:
        Bus_Names_APSRTC.append(bus.text)
        Route_Links.append(link_APSRTC)
        Route_names.append(Route_APSRTC)
    for bus_types in Bus_Types:
        Bus_Types_APSRTC.append(bus_types.text)
    for departure_time in Departure_Time:
        Departure_Time_APSRTC.append(departure_time.text)
    for destination_time in Destination_Time:
        Destination_Time_APSRTC.append(destination_time.text)
    for total_duration in Total_Duration:   
        Total_Duration_APSRTC.append(total_duration.text)
    for ratings in Ratings:
        Ratings_APSRTC.append(ratings.text)
    for price in Prices:
        Prices_APSRTC.append(price.text)
    for seat in Seats_Available:    
        Seats_Available_APSRTC.append(seat.text)


print("Successfully Completed")
# Close the browser
driver_APSRTC.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time

# Retrieve the bus details 
driver_TSRTC = webdriver.Chrome()
driver_TSRTC.maximize_window()
Bus_Names_TSRTC = []
Bus_Types_TSRTC = []
Departure_Time_TSRTC = []
Destination_Time_TSRTC = []
Total_Duration_TSRTC = []
Ratings_TSRTC = []
Prices_TSRTC = []
Seats_Available_TSRTC = []
Route_names = []
Route_Links = []

for i, r in df.iterrows():
    link_TSRTC = r["Bus Routes Link"]
    Route_TSRTC = r["Bus Routes Name"]

    # Navigate to the link
    driver_TSRTC.get(link_TSRTC)
    time.sleep(2)

    # Loop to handle multiple buttons on the page (top, center, bottom)
    buttons_clicked = 0
    while True:
        buttons_to_click = driver_TSRTC.find_elements(By.XPATH, "//div[@class='button']")
        
        if buttons_to_click and buttons_clicked < len(buttons_to_click):
            try:
                # Click the next button
                buttons_to_click[buttons_clicked].click()
                buttons_clicked += 1
                time.sleep(2)
                
                # Scroll slowly after each click to reveal the next button
                for _ in range(3):  # Scroll in small increments
                    ActionChains(driver_TSRTC).send_keys(Keys.PAGE_DOWN).perform()
                    time.sleep(1)

            except Exception as e:
                print(f"Error clicking button: {e}")
                continue
        else:
            print("No more buttons to click or none found on the page, proceeding to extract details.")
            break

    # Scroll the entire page slowly to ensure all details are loaded
    scrolling = True
    while scrolling:
        old_page_source = driver_TSRTC.page_source

        # Use Actions to perform a slow PAGE_DOWN
        for _ in range(3):  # Scroll in small increments
            ActionChains(driver_TSRTC).send_keys(Keys.PAGE_DOWN).perform()
            time.sleep(1)

        new_page_source = driver_TSRTC.page_source

        if new_page_source == old_page_source:
            scrolling = False

    # Extract bus details
    Bus_Names = driver_TSRTC.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    Bus_Types = driver_TSRTC.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    Departure_Time = driver_TSRTC.find_elements(By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']")
    Destination_Time = driver_TSRTC.find_elements(By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']")
    Total_Duration = driver_TSRTC.find_elements(By.XPATH, "//div[@class='dur l-color lh-24']")
    
    Ratings = driver_TSRTC.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    Prices = driver_TSRTC.find_elements(By.XPATH, "//div[@class='fare d-block']")
    Seats_Available = driver_TSRTC.find_elements(By.XPATH, "//div[contains(@class,'seat-left')]")

    # Append data to respective lists
    for bus in Bus_Names:
        Bus_Names_TSRTC.append(bus.text)
        Route_Links.append(link_TSRTC)
        Route_names.append(Route_TSRTC)
    for bus_types in Bus_Types:
        Bus_Types_TSRTC.append(bus_types.text)
    for departure_time in Departure_Time:
        Departure_Time_TSRTC.append(departure_time.text)
    for destination_time in Destination_Time:
        Destination_Time_TSRTC.append(destination_time.text)
    for total_duration in Total_Duration:   
        Total_Duration_TSRTC.append(total_duration.text)
    for ratings in Ratings:
        Ratings_TSRTC.append(ratings.text)
    for price in Prices:
        Prices_TSRTC.append(price.text)
    for seat in Seats_Available:    
        Seats_Available_TSRTC.append(seat.text)

print("Successfully Completed")
# Close the browser
driver_TSRTC.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time

# Retrieve the bus details 
driver_KSRTC_Kerala = webdriver.Chrome()
driver_KSRTC_Kerala.maximize_window()
Bus_Names_KSRTC_Kerala = []
Bus_Types_KSRTC_Kerala = []
Departure_Time_KSRTC_Kerala = []
Destination_Time_KSRTC_Kerala = []
Total_Duration_KSRTC_Kerala = []
Ratings_KSRTC_Kerala = []
Prices_KSRTC_Kerala = []
Seats_Available_KSRTC_Kerala = []
Route_names = []
Route_Links = []

for i, r in df.iterrows():
    link_KSRTC = r["Bus Routes Link"]
    Route_KSRTC = r["Bus Routes Name"]

    # Navigate to the link
    driver_KSRTC_Kerala.get(link_KSRTC)
    time.sleep(2)

    # Loop to handle multiple buttons on the page (top, center, bottom)
    buttons_clicked = 0
    while True:
        buttons_to_click = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[@class='button']")
        
        if buttons_to_click and buttons_clicked < len(buttons_to_click):
            try:
                # Click the next button
                buttons_to_click[buttons_clicked].click()
                buttons_clicked += 1
                time.sleep(2)
                
                # Scroll slowly after each click to reveal the next button
                for _ in range(3):  # Scroll in small increments
                    ActionChains(driver_KSRTC_Kerala).send_keys(Keys.PAGE_DOWN).perform()
                    time.sleep(1)

            except Exception as e:
                print(f"Error clicking button: {e}")
                continue
        else:
            print("No more buttons to click or none found on the page, proceeding to extract details.")
            break

    # Scroll the entire page slowly to ensure all details are loaded
    scrolling = True
    while scrolling:
        old_page_source = driver_KSRTC_Kerala.page_source

        # Use Actions to perform a slow PAGE_DOWN
        for _ in range(3):  # Scroll in small increments
            ActionChains(driver_KSRTC_Kerala).send_keys(Keys.PAGE_DOWN).perform()
            time.sleep(1)

        new_page_source = driver_KSRTC_Kerala.page_source

        if new_page_source == old_page_source:
            scrolling = False

    # Extract bus details
    Bus_Names = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    Bus_Types = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    Departure_Time = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[@class='dp-time f-19 d-color f-bold']")
    Destination_Time = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[@class='bp-time f-19 d-color disp-Inline']")
    Total_Duration = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[@class='dur l-color lh-24']")
    
    Ratings = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    Prices = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[@class='fare d-block']")
    Seats_Available = driver_KSRTC_Kerala.find_elements(By.XPATH, "//div[contains(@class,'seat-left')]")

    # Append data to respective lists
    for bus in Bus_Names:
        Bus_Names_KSRTC_Kerala.append(bus.text)
        Route_Links.append(link_KSRTC)
        Route_names.append(Route_KSRTC)
    for bus_types in Bus_Types:
        Bus_Types_KSRTC_Kerala.append(bus_types.text)
    for departure_time in Departure_Time:
        Departure_Time_KSRTC_Kerala.append(departure_time.text)
    for destination_time in Destination_Time:
        Destination_Time_KSRTC_Kerala.append(destination_time.text)
    for total_duration in Total_Duration:   
        Total_Duration_KSRTC_Kerala.append(total_duration.text)
    for ratings in Ratings:
        Ratings_KSRTC_Kerala.append(ratings.text)
    for price in Prices:
        Prices_KSRTC_Kerala.append(price.text)
    for seat in Seats_Available:    
        Seats_Available_KSRTC_Kerala.append(seat.text)

print("Successfully Completed")
# Close the browser
driver_KSRTC_Kerala.quit()
```
